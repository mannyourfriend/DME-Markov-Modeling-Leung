```python
cost1 = costsDict['cost1'].replace("]", '")').replace("[", '("').lower()
cost2 = costsDict['cost2'].replace("]", '")').replace("[", '("').lower()
cost3 = costsDict['cost3'].replace("]", '")').replace("[", '("').lower()
cost4 = costsDict['cost4'].replace("]", '")').replace("[", '("').lower()
cost5 = costsDict['cost5'].replace("]", '")').replace("[", '("').lower()
cost6 = costsDict['cost6'].replace("]", '")').replace("[", '("').lower()
cost7 = costsDict['cost7'].replace("]", '")').replace("[", '("').lower()
cost8 = costsDict['cost8'].replace("]", '")').replace("[", '("').lower()
cost9 = costsDict['cost9'].replace("]", '")').replace("[", '("').lower()
cost10 = costsDict['cost10'].replace("]", '")').replace("[", '("').lower()
# def evaluate_expression(expression, stAge, ptAge, ctAge):
#     global _stage, ptage, caretakerage
#     _stage = stAge
#     ptage = ptAge
#     caretakerage = ctAge
#     if (isinstance(expression, (float,int))):
#         return expression
#     # Fetch all variable names from the expression using regex
#     variables = re.findall(r'\b[a-zA-Z_][a-zA-Z0-9_]*\b', expression)
#     unique_vars = set(variables)
#     # Check each variable, if it's not defined or is None, set it using get_average_value
#     for var in unique_vars:
#         tempVal = get_rand_value(var)
#         if tempVal is not None:
#             globals()[var] = tempVal
#     if (isinstance(expression, str)) and re.search(r'\bcost\d+\b', expression):
#         theEval = eval(expression, {"__builtins__": None}, globals())
#         print(theEval)
#         return(evaluate_expression(theEval, stAge, ptAge, ctAge))
#     return eval(expression, {"__builtins__": None}, globals())

def evaluate_expression(expression, stAge, ptAge, ctAge):
    global _stage, ptage, caretakerage
    _stage = stAge
    ptage = ptAge
    caretakerage = ctAge
    if (isinstance(expression, int) or isinstance(expression, float)):
        return expression
    # Fetch all variable names from the expression using regex
    variables = re.findall(r'\b[a-zA-Z_][a-zA-Z0-9_]*\b', expression)
    unique_vars = set(variables)
    # Check each variable, if it's not defined or is None, set it using get_average_value
    for var in unique_vars:
        tempVal = get_rand_value(var)
        if tempVal is not None:
            globals()[var] = get_rand_value(var)
    if (isinstance(expression, str)):
        if len(re.findall(r'\bcost\d+\b', expression)) > 0:
            theEval = eval(expression, {"__builtins__": None}, globals())
            return(evaluate_expression(theEval, stAge, ptAge, ctAge))
    # Evaluate the expression safely
    return eval(expression, {"__builtins__": None}, globals())

def get_rand_value(variable_name, df = tableVarCosts):
    """
    Searches for a variable in the 'Variable' column of the DataFrame and returns
    the corresponding 'Average Value'.
    
    Parameters:
        variable_name (str): The name of the variable to look up.
        df (pd.DataFrame): The DataFrame containing the variable data.

    Returns:
        float: The average value of the variable or None if not found.
    """
    # Filter the DataFrame to find the row where the 'Variable' column matches the input string
    matched_row = df[df['Variable'] == variable_name]
    # Check if there's a match and return the 'Average Value' or None
    if not matched_row.empty:
        above = random.random() > 0.5
        min = matched_row['Minimum Value'].iloc[0]
        max = matched_row['Maximum Value'].iloc[0]
        avg = matched_row['Average Value'].iloc[0]
        if isinstance(min, int) and isinstance(max, int) and isinstance(avg, int):
            if above:
                return random.randrange(avg, max)
            else:
                return random.randrange(min, avg)
        else:
            if above:
                return random.uniform(avg, max)
            else:
                return random.uniform(min, avg)
    else:
        return None

    
def tablept(column):
    if ptage > 75:
        # return 0
        matched_row = tablePT[tablePT['Age'] == 75] 
    else:
        matched_row = tablePT[tablePT['Age'] == ptage]
    if not matched_row.empty:
        return matched_row[column].iloc[0]
    else:
        return None
    
def tablecaretaker(column):
    if caretakerage > 70:
        # return 0
        matched_row = tableCARETAKERAGE[tableCARETAKERAGE['Age'] == 70]
    else:
        matched_row = tableCARETAKERAGE[tableCARETAKERAGE['Age'] == caretakerage]
    if not matched_row.empty:
        return matched_row[column].iloc[0]
    else:
        return None
def tableswitch():
    # return 0
    tempstage = _stage
    if tempstage > 17:
        tempstage = 17
    matched_row = tableSWITCH[tableSWITCH['index'] == tempstage]
    if not matched_row.empty:
        return matched_row["value"].iloc[0]
    else:
        return None
def tabledie():
    age_ranges = [
        (0, 61, 0),
        (61, 65, 0.039349805),
        (65, 70, 0.050723952),
        (70, 75, 0.052984008),
        (75, 120, 0.0496836),
        (120, float('inf'), 1)  # Assumes ptage will not exceed 120 for practical purposes
    ]
    
    # Iterate over the ranges to find where ptage fits
    for start, end, RoD in age_ranges:
        if start <= ptage < end:
            return RoD  # Return the RoD value for the matched range
    
    # If no range is matched, which shouldn't happen, return None or an appropriate value
    return None
```
